```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

filepath = 'D:\\algo_data_science\\Global_AI_Content_Impact_Dataset.csv'
df = pd.read_csv(filepath)
df.head()
```

```python
indonesia_data = df[df['Country'] == 'Indonesia']
if not indonesia_data.empty:
    indonesia_data 
else:
    print("Tidak ada data untuk Indonesia.")
```

```python
media_data = df[df['Industry'] == 'Media']

avg_adoption_by_year = media_data.groupby('Year')['AI Adoption Rate (%)'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(avg_adoption_by_year['Year'], avg_adoption_by_year['AI Adoption Rate (%)'], marker='o', color='teal', linewidth=2)
plt.title('Tren Rata-rata Adopsi AI di Industri Media Global (2020-2025)')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata AI Adoption Rate (%)')
plt.grid(True)
plt.xticks(avg_adoption_by_year['Year'])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10), facecolor='lightblue')

sns.histplot(data=media_data, x='Year', weights='AI-Generated Content Volume (TBs per year)', bins=6, color='teal', ax=axes[0, 0])
axes[0, 0].set_title('Distribusi Volume Konten AI di Industri Media per Tahun')
axes[0, 0].set_xlabel('Tahun')
axes[0, 0].set_ylabel('Total Volume Konten AI (TBs)')

sns.boxplot(data=media_data, x='Year', y='AI-Generated Content Volume (TBs per year)', color='teal', ax=axes[0, 1])
axes[0, 1].set_title('Boxplot Volume Konten AI di Industri Media per Tahun')
axes[0, 1].set_xlabel('Tahun')
axes[0, 1].set_ylabel('Volume Konten AI (TBs per year)')

sns.histplot(data=media_data, x='Regulation Status', weights='AI Adoption Rate (%)', color='teal', ax=axes[1, 0])
axes[1, 0].set_title('Distribusi Tingkat Adopsi AI Berdasarkan Status Regulasi')
axes[1, 0].set_xlabel('Status Regulasi')
axes[1, 0].set_ylabel('Total Tingkat Adopsi AI (%)')

sns.boxplot(data=media_data, x='Regulation Status', y='AI Adoption Rate (%)', color='teal', ax=axes[1, 1])
axes[1, 1].set_title('Boxplot Tingkat Adopsi AI Berdasarkan Status Regulasi')
axes[1, 1].set_xlabel('Status Regulasi')
axes[1, 1].set_ylabel('Tingkat Adopsi AI (%)')

plt.tight_layout()

plt.show()
```
